```python
from PIL import Image

W = 10
H = 10
maxV = 4
step = 1
img = Image.new('RGB', [W, H], maxV)
data = img.load()

for x in range(img.size[0]):
    for y in range(img.size[1]):
        data[x,y] = (
            0,
            0,
            0,
        )
def DoNext(data):
    for x in range(W):
       for y in range(H):
            sum = data[x,y][0] + maxV * data[x,y][1] + maxV**2 * data[x,y][2]
            sum += step
            maximum = (maxV)**3
            if sum >= maximum:
                data[x,y] = (0, 0, 0)
            else:
                data[x,y] = (sum % maxV, (sum // maxV) % maxV * step, (sum // maxV // maxV) % maxV * step)
                print(str(data[x,y]) + ",")
                return data

    return data

for i in range(maxV **3):
    data = DoNext(data)
```
